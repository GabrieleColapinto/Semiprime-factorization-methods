```python
from matplotlib import pyplot as plt

N = 91 # Product of prime numbers
a = 83 # Base for the period finding
r = 4 # Period of the function

# Calculation of the data for the plot
max_x = 13
xvals = range(max_x)
yvals = [(a**x)%N for x in xvals]

# Information of the plot
plt.plot(xvals, yvals, linewidth=1, linestyle='dotted', marker='x')
plt.xlabel('$x$')
plt.ylabel(f'${a}^x$ mod ${N}$')
plt.title("Example of Periodic Function in Shor's Algorithm")

# Annotation of the period of the function
plt.annotate('', xy=(0,1), xytext=(r,1),
                 arrowprops=dict(arrowstyle='<->'))
plt.annotate(f'$r={r}$', xy=(r/3,1.5))

plt.show()
```
