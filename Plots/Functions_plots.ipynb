```python
import sympy as sp
from sympy import exp, log, sqrt
from numpy import linspace
from matplotlib import pyplot as plt
```

```python
# We define the time complexities of the algorithms as functions
n = sp.Symbol('n')
brute_force = (2**(n/2))*n**2
fermat_factorization = (2**n - 2**(n/2))*(n**2 + n)
quadratic_sieve = exp(sqrt(log(n)*log(log(n))))
shor = (n**2)*log(n)*(log(log(n)))
```

```python
# We define the range of the number of bits of the number to be factored
min_bits = 10
max_bits = 256
samples = 1000
endpoint = True

# We define the values of the algorithms
xvals = linspace(min_bits, max_bits, samples, endpoint)

yvals_bf = []
yvals_ff = []
yvals_qs = []
yvals_sh = []

for x in xvals:
    yvals_bf.append(brute_force.evalf(subs={n: x}))
    yvals_ff.append(fermat_factorization.evalf(subs={n: x}))
    yvals_qs.append(quadratic_sieve.evalf(subs={n: x}))
    yvals_sh.append(shor.evalf(subs={n: x}))
```

```python
# We plot the data
plt.plot(xvals, yvals_bf, linewidth=1, color='r')
plt.xlabel('Number of bits of the input: $n$')
plt.ylabel('Time complexity of the algorithm')
plt.title("Brute force algorithm")
#plt.savefig("Brute_force_plot.jpg")
plt.show()
```

```python
# We plot the data
plt.plot(xvals, yvals_ff, linewidth=1, color='g')
plt.xlabel('Number of bits of the input: $n$')
plt.ylabel('Time complexity of the algorithm')
plt.title("Fermat's factorization algorithm")
#plt.savefig("Fermat_factorization_plot.jpg")
plt.show()
```

```python
# We plot the data
plt.plot(xvals, yvals_qs, linewidth=1, color='b')
plt.xlabel('Number of bits of the input: $n$')
plt.ylabel('Time complexity of the algorithm')
plt.title("Quadratic sieve algorithm")
#plt.savefig("Quadratic_sieve_plot.jpg")
plt.show()
```

```python
# We plot the data
plt.plot(xvals, yvals_sh, linewidth=1, color='y')
plt.xlabel('Number of bits of the input: $n$')
plt.ylabel('Time complexity of the algorithm')
plt.title("Shor's algorithm")
#plt.subplots_adjust(left=0.15)
#plt.savefig("Shor_algorithm_plot.jpg")
plt.show()
```
